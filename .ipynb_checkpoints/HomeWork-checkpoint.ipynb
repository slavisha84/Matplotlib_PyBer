# Summary
1. The largerst concentration of drivers is located in urban cities ~%81). 
2. Aproximetaly %30 of fares are in suburban locations.
3. Urban areas have only %2.6 or drivers who are responsible for ~%5 of total rides. 

```python
%matplotlib inline
# Importing Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)
```

```python
# Exmploring city data set
city_df.head()
```

```python
# Exploring ride data set
ride_df.head()
```

```python
# Combine the data into a single dataset using left join
all_df = pd.merge(ride_df, city_df, on = "city", how = "left")
# Display the data table for preview
all_df.head()
```

```python
###

# Create separate data frames based on city types
urban = all_df.loc[(all_df["type"] == "Urban")]
suburban = all_df.loc[(all_df["type"] == "Suburban")]
rural = all_df.loc[(all_df["type"] == "Rural")]
```

```python
# Urban cities - calculations of Average Fare, Total Rides and Total Drivers
avg_fare_urban = urban.groupby(["city"])["fare"].mean()
total_rides_urban = urban.groupby(["city"])["ride_id"].count()
total_drivers_urban = urban.groupby(["city"])["driver_count"].sum()

# Suburban cities - calculations of Average Fare, Total Rides and Total Drivers
avg_fare_suburban = suburban.groupby(["city"])["fare"].mean()
total_rides_suburban = suburban.groupby(["city"])["ride_id"].count()
total_drivers_suburban = suburban.groupby(["city"])["driver_count"].sum()

# Rural cities - calculations of Average Fare, Total Rides and Total Drivers
avg_fare_rural = rural.groupby(["city"])["fare"].mean()
total_rides_rural = rural.groupby(["city"])["ride_id"].count()
total_drivers_rural = rural.groupby(["city"])["driver_count"].sum()
```

```python
# Create scatter plots for Urban, Suburban and Rural cities
plt.scatter(total_rides_urban, avg_fare_urban,s = total_drivers_urban,
            marker ='o', facecolors ="lightcoral", edgecolors='black',alpha = 0.5, label="Urban", )

plt.scatter(total_rides_suburban, avg_fare_suburban,s = total_drivers_suburban,
            marker ='o', facecolors ="lightskyblue", edgecolors='black',alpha = 0.5, label="Suburban")

plt.scatter(total_rides_rural, avg_fare_rural,s=total_drivers_rural,
            marker ='o', facecolors ="gold", edgecolors='black',alpha = 0.55, label="Rural")

# Chart title
plt.title("Pyber Ride Sharing Data (2016)")

# x label
plt.xlabel("Total Numbers of Rides (Per City))")

# y label
plt.ylabel("Average Fare($)")

# Note
plt.text(42,35,"Note:\nCircle size correlates with number of drivers per city")

# legend 
lgnd = plt.legend(frameon = True, edgecolor = 'black')
lgnd.legendHandles[0]._sizes = [75]
lgnd.legendHandles[1]._sizes = [75]
lgnd.legendHandles[2]._sizes = [75]

# Save the image of the chart and print on the screen
plt.savefig("Scatter.png")
plt.show()
```

```python
### Calculate Type Percents of fare by city type

# Isolating columns to work with
fbc_df = all_df.loc[:,["type","fare"]]

# Grouping by city and summing by fare
fbc_group = fbc_df.groupby("type")["fare"].sum()

# Previewing data
fbc_group.head()
```

```python
# I will be building the elements for pie chart one by one in this example
# Setting the lables of the pie chart
labels = ["Rural","Suburban","Urban"]

# Setting the colors
colors = ["gold","lightskyblue","lightcoral"]

# Setting pie chart to explode mode
explode = (0, 0.1, 0.1)

# Setting the title
plt.title("% of Total Fares By City Types")

# Ploting the data
plt.pie(fbc_group, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%",shadow=True, startangle=160)
plt.axis("equal")

# Saving the image and showing the plot
plt.savefig("%F_by_C.png")
plt.show()
```

```python
### Total Riders by CIty Type

rbc_df = all_df.loc[:,["type","ride_id"]]
rbc_group = rbc_df.groupby("type")["ride_id"].count()
rbc_group.head()

# Setting the title
plt.title("% of Total Rides By City Types")

# Ploting the data with less line of code comparing to the previous pie
plt.pie(rbc_group, explode=(0, 0.1, 0.1), labels=["Rural","Suburban","Urban"], colors=["gold","lightskyblue","lightcoral"], autopct="%1.1f%%",shadow=True, startangle=160)
plt.axis("equal")

# Saving the image and showing the plot
plt.savefig("%R_by_C.png")
plt.show()
```

```python
### TOTAL DRIVERS BY CITY TYPE

# to reduce cleaning effort i will use city_data before i merged them since i dont need anything else
dpc_group =city_df.groupby("type")["driver_count"].sum()

# Setting the title
plt.title("% of Total Drivers By City Types")

# Ploting the data with less line of code comparing to the previous pie
plt.pie(dpc_group, explode=(0, 0.1, 0.1), labels=["Rural","Suburban","Urban"], colors=["gold","lightskyblue","lightcoral"], autopct="%1.1f%%",shadow=True, startangle=160)
plt.axis("equal")

# Saving the image and showing the plot
plt.savefig("%D_by_C.png")
plt.show()
```
